```python
import numpy as np
import xarray as xr
import glob
import os
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
# Supply dir path and read data

ppdir = "/mnt/storage6/myers/NEMO/ANHA4-EPM111/"

ppdir_grid = "/mnt/storage6/myers/NEMO_meshes/ANHA4/50VerticalLevels/"

ds = xr.open_mfdataset(ppdir + "*1958*gridT.nc")
ds_ice = xr.open_mfdataset(ppdir + "*1958*icemod.nc")

ds_grid = xr.open_dataset(ppdir_grid + "mesh_mask_anha4.nc")

ds = xr.merge([ds.rename({'y_grid_T':'y', 'x_grid_T':'x', 'deptht':'z', 'time_counter':'time', 
                         'nav_lat_grid_T':'lat', 'nav_lon_grid_T':'lon'}),
               ds_ice.rename({'time_counter':'time'}).drop(['nav_lat', 'nav_lon']),
               ds_grid.get(['e1t', 'e2t', 'tmask', 'nav_lat', 'nav_lon'])])
```

```python
## testing volume-integration

cell_vol = ds['e2t'] * ds['e1t'] * ds['e3t']
cell_vol = cell_vol.where(ds['tmask'] > 0)

cell_area = ds['e2t'] * ds['e1t']
cell_area = cell_area.where(ds['tmask'].isel(z=0) > 0) 

# subpolar North Atlantic mask
dV = cell_vol.where((ds['lat']>=45.) & (ds['lat']<=70.))

theta = (ds['votemper'] * dV).sum(['x', 'y'])

sea_ice_vol = (ds['ileadfra'] * cell_area * ds['iicethic']).sum(['x', 'y'])
```

## Testing ouput


```python
ppdir = "/mnt/storage6/hemant/Memory-ANHA4-EPM111/"

ds1 = xr.open_mfdataset(ppdir + "Ocean_temp*nc")

ds1 = ds1.sel(z=slice(0.,500.)).sum('z')
```

```python
ds1
```

```python
fig = plt.figure(figsize=[14,4])

tim = ds1['time'].dt.year + ds1['time'].dt.month / 12 + ds1['time'].dt.day / 365.

plt.subplot(1,3,1)
plt.plot(tim, ds1['votemper_subpolar_NAtl'])
plt.plot(tim, ds1['votemper_arctic'])
plt.title('Oean heat content (upper 500 m)')

plt.subplot(1,3,2)
plt.plot(tim, ds1['sohefldo_subpolar_NAtl'])
plt.plot(tim, ds1['sohefldo_arctic'])
plt.title('Heat flux into the ocean')

plt.subplot(1,3,3)
plt.plot(tim, ds1['ileadfra'])
plt.plot(tim, ds1['ice_vol'])
plt.title('Sea ice area and volume')
```
